```python
import re

exceptions = ['ch', 'wh', 'kn', 'gn' 'th', 'sh', 'www', 'wr', 'ph', 'htt']

l = []
final_list = []

for i in al_tok:
    res = re.search('^[bcdfghjklmnpqrstvwxz]{2,3}', i)
    
    if res:
        l.append(res.group())
        
for c in l:
    if c in exceptions:
        continue
    else:
        final_list.append(c)
    
    
print(final_list)
        


```

```python
cl_dict = {}
    
for element in final_list:
    cl_dict[element] = cl_dict.get(element, 0) + 1 


print(cl_dict)
```

```python
print('Total number of clusters: ', len(final_list))
print('Number of different clusters: ', len(cl_dict))
```

```python
sorted_list = []

for key, value in sorted(cl_dict.items()):
    sorted_list.append(key)
    
print(sorted_list)
```
